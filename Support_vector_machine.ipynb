Linear svm implementation uding sklearn

```python
import numpy as np  
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
```

```python
iris=datasets.load_iris()
```

```python
print(iris.keys())
```

```python
x=iris["data"][:,(2,3)]
y=(iris["target"]==2).astype(float)
```

```python
svm_clf=Pipeline([
    ("scaler",StandardScaler()),
    ("linear_svc",LinearSVC(C=1,loss="hinge")),
])
# The following lines of code would also give the same output but no pipelines.
# svm_clf=LinearSVC(C=1,loss="hinge")
```

```python
svm_clf.fit(x,y)
```

```python
svm_clf.predict([[5.5,1.7]])
```

Non linear svm using sklearn

```python

from sklearn.datasets import make_moons
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
```

```python
moon=make_moons()
print(moon)
```

```python
print(moon[0])
```

```python
polynomial_svm_clf=Pipeline([
    ("polynomial_features",PolynomialFeatures(degree=3)),
    ("sclaer",StandardScaler()),
    ("svm_clf",LinearSVC(C=10,loss="hinge"))
])
```

```python
polynomial_svm_clf.fit(x, y)
```

```python
print(polynomial_svm_clf.predict([[2.5,1.5]]))
```
