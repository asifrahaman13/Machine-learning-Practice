```python
import numpy as np 

#Setting the data points manually using numpy.
x=2*np.random.rand(100,1)
# y = 4 + 3 * x + np.random.randint(1, 100,1)
#If we use the following then we will get number changing every time we restart the program.
y = 4 + 3 * x + np.random.randn(100,1)

#ridge regression.
from sklearn.linear_model import Ridge
regression=Ridge(alpha=1,solver="cholesky")
regression.fit(x,y)
print(f"The ridge regression predicted the following: {regression.predict([[1.5]])}")

#sgd regression. 
from sklearn.linear_model import SGDRegressor
sgd_regression=SGDRegressor(penalty="l2")
sgd_regression.fit(x,y.ravel())
#The ravel() function will add only one dimension of the arrays even if there are multiple dimensions.
print(f"The sgd regression predicted the following: {sgd_regression.predict([[1.5]])}")

#lasso regression.
from sklearn.linear_model import Lasso
lass_regression=Lasso(alpha=0.1)
lass_regression.fit(x,y)
print(f"The lasso regression predicted the following: {lass_regression.predict([[1.5]])}")

#Elastic net regression.
from sklearn.linear_model import ElasticNet
elastic_regression=ElasticNet(alpha=0.1,l1_ratio=0.5)
elastic_regression.fit(x,y)
print(f"The elastic net regression predicted the following: {elastic_regression.predict([[1.5]])}")
```
