DBSCAN will look for some closeness of the clusters in the figure and then it will look if the points lies within the clusters.If they lies within it then they are considered in the same classes.

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
x,y=make_moons(n_samples=1000,noise=0.05)
dbscan=DBSCAN(eps=0.05,min_samples=5)
dbscan.fit(x)
print(dbscan.labels_)
```

```python
print(dbscan.core_sample_indices_)
dbscan.core_sample_indices_
dbscan.components_
```
