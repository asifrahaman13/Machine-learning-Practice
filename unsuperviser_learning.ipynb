```python
import numpy as np  
import matplotlib.pyplot as plt   
```

```python
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
style.use("fivethirtyeight")
x, y = make_blobs(n_samples = 100, centers = 3,cluster_std = 1, n_features = 2)
```

```python
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x[:, 0], x[:, 1], s = 40, color = 'b')
plt.show()
plt.clf()
```

```python
from sklearn.cluster import KMeans
#The following two lines of codes will make sure to get the best possible initialization of the centroid.
good_init=np.array([[-1,-7],[1,9],[1,-6]])
kmeans=KMeans(n_clusters=3,init=good_init,n_init=1)
kmeans.fit_transform(x,y)
```

```python
y_pred=kmeans.predict(x)
print(y_pred)
```

```python
print(kmeans.cluster_centers_)
```

```python
x_new=np.array([[0,2],[3,2],[-3,3]])
kmeans.predict(x_new)
```

We can predict whether the prediction is correct or not with the help of silhouette function.
1.1 means that the prediction is correct.
2.0 means that the prediction is close to the boundary.
3.-1 means that the prediction is probably in the wrong cluster.

```python
from sklearn.metrics import silhouette_score
silhouette_score(x,kmeans.labels_)
```
