```python
import numpy as np  
import matplotlib.pyplot as plt   
```

```python
from sklearn.datasets import make_blobs
from matplotlib import pyplot as plt
from matplotlib import style
```

```python
style.use("fivethirtyeight")
x, y = make_blobs(n_samples = 100, centers = 3,cluster_std = 1, n_features = 2)
```

```python
plt.xlabel("x")
plt.ylabel("y")
plt.scatter(x[:, 0], x[:, 1], s = 40, color = 'b')
plt.show()
plt.clf()
```

```python
from sklearn.cluster import KMeans
#The following two lines of codes will make sure to get the best possible initialization of the centroid.
good_init=np.array([[-1,-7],[1,9],[1,-6]])
kmeans=KMeans(n_clusters=3,init=good_init,n_init=1)
kmeans.fit(x,y)
```

```python
y_pred=kmeans.predict(x)
print(y_pred)
```

```python
print(kmeans.labels_)
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.inertia_)
#Inertia is not a good estimation for finding the optimal solution for the value of k.This is because as the number of instances or data points grow in th datasets the distance between the centroids and the data points becomes closer and closer.
```

```python
print(kmeans.score(x))
```

For improving the system performance and to consume lesser memory the sklearn will use the technique of mini batch clustering algorithms.

```python
from sklearn.cluster import MiniBatchKMeans
minibatch_kmeans=MiniBatchKMeans(n_clusters=5)
minibatch_kmeans.fit(x)
```

```python
print(minibatch_kmeans.predict(x))
```

```python
#The silhouette score helps to determine the number off score.
from sklearn.metrics import silhouette_score
score=silhouette_score(x,kmeans.labels_)
print(score)
```
