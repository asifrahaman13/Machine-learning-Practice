```python
import pandas as pd  
import matplotlib.pyplot as plt
import numpy as np   
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix,classification_report
data=pd.read_csv('datas/creditcard.csv')
print(data.info())
```

```python
scalar=StandardScaler()
data['normAmount']=scalar.fit_transform(np.array(data['Amount']).reshape(-1,1))
data=data.drop(['Time', 'Amount'], axis = 1)
data['Class'].value_counts()
```

```python
df=pd.DataFrame(data)
x=df["V1"]
y=df["Class"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=9)
print(f"No of transactions in x_train dataset is: {x_train.shape}")
print(f"No of transactions in x_test dataset is: {x_test.shape}")
print(f"No of transactions in y_train dataset is: {y_train.shape}")
print(f"No of transactions in y_test dataset is: {y_test.shape}")
```

```python
regression=LogisticRegression()
```

```python
regression.fit(x_train,y_train)
predictions=regression.predict([x_test])
print(classification_report(y_test,predictions))
```
