Decision tree classification using sklearn

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
```

```python
iris=load_iris()
print(iris)
```

```python
x=iris.data[:,2:]
# print(x)
```

```python
y=iris.target
print(y)
```

```python
tree_regression=DecisionTreeClassifier(max_depth=2)
#The max depth is used to specify how many exactly segments needed to be made in the classification tree.
```

```python
tree_regression.fit(x,y)
```

```python
print(tree_regression.predict([[2.4,5]]))
```

```python
print(tree_regression.predict_proba([[2.4,5]]))
#The training set will return that class which will be having the heighest probability among all the three.Here the third class have the heighest probability so is the answer as per the model.
```

Decision tree regression technique

```python
from sklearn.tree import DecisionTreeRegressor
import numpy as np
```

```python
x=np.arange(1,100,1)
print(x)
```

```python
y=[np.power(x,2)+3]
print(y)
```

```python
tree_classification_regreesor=DecisionTreeRegressor(max_depth=2)
```

```python
tree_classification_regreesor.fit([x],y)
```

```python
print(tree_classification_regreesor.predict([[int(x) for x in range(101,200)]]))
print(tree_classification_regreesor)
```
