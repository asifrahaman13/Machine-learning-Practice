Ensemble learning is the process in which the model will predict the output based on the output produced by different classification techniques in the ptogram.The majority vote classifier is known as the hard voting classifier.

Even if there are all the learners which are weak but still the prediction will be upto a good accuracy model iff there are a number of learning algorithms and the learning algorithms are quite diverse from one another.

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC 
```

```python
logistic_regression=LogisticRegression()
random_forest_regression=RandomForestClassifier()
svc_regression=SVC()
```

```python
iris=load_iris()
print(iris.keys())
# print(iris)
x=iris.data[:,2:]
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
voting_clf = VotingClassifier(
    estimators=[('lr', logistic_regression), ('rf', random_forest_regression), ('svc', svc_regression)],
    voting='hard')
    #The soft voting is even better than the hard voting system.
voting_clf.fit(x_train, y_train)
```

```python
print(voting_clf.predict([[0.5,8.9]]))
```

```python
from sklearn.metrics import accuracy_score
for clf in (logistic_regression,random_forest_regression,svc_regression,voting_clf):
    clf.fit(x_train,y_train)
    y_pred = clf.predict(x_test)
    print(f"The accuracy score of {clf} is: {accuracy_score(y_test,y_pred)}")
```

Bagging and pasting implementation

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
bag_clf=BaggingClassifier(
    DecisionTreeClassifier(),n_estimators=500,max_samples=100,bootstrap=True,n_jobs=-1,oob_score=True
)
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
print(iris.keys())
# print(iris)
x=iris.data[:,2:]
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
bag_clf.fit(x_train, y_train)
```

```python
y_pred=bag_clf.predict(x_test)
```

```python
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
print(bag_clf.oob_score_)
#We can compare the accuracy score with the oob score and it will always be very close enough with each other.
```
