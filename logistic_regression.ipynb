```python
import numpy as np   
import matplotlib.pyplot as plt   
x=np.arange(-10,10,0.1)
y=1/(1+np.exp(-1*x))
plt.plot(x,y,color="purple")
plt.xlabel("The x axis")
plt.ylabel("The y axis")
plt.legend(["logistic regression graph"])
plt.show()
```

Logistic regression using the iris data sets.

```python
from sklearn import datasets
iris=datasets.load_iris()
print(iris.keys())
```

```python
x=iris["data"][:,3:]
y=(iris["target"]==2).astype(int) 
#The above code will make sure to display the code in the form of integers.
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
log_reg.fit(x,y)
```

```python
import matplotlib.pyplot as plt
X_new = np.arange(0, 3, 0.00001).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
plt.plot(X_new, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", label="Not Iris-Virginica")
plt.legend(["iris-verginia","non iris verginia"])
plt.show()
```
